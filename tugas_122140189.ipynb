```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat Array T (waktu) dari 0 sampai 2 dengan langkah sebesar 0.0001
T = np.arange(0, 2, 0.0001)

# Perhitungan Sinyal
y1 = 2 * np.sin(2 * np.pi * 3 * T + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * T + (np.pi/4))
y3 = -1 * np.sin(2 * np.pi * 5 * T + (np.pi/2))
y4 = 0.5 * np.cos(2 * np.pi * 6 * T + np.pi)

# Membuat Subplot
plt.figure(figsize=(12, 8))

plt.subplot(4,1,1)
plt.plot(T, y1)
plt.title('y1 = 2sin(2π3t)')
plt.grid(True)

plt.subplot(4,1,2)
plt.plot(T, y2)
plt.title('y2 = cos(2π4t + π/4)')
plt.grid(True)

plt.subplot(4,1,3)
plt.plot(T, y3)
plt.title('y3 = -sin(2π5t + π/2)')
plt.grid(True)

plt.subplot(4,1,4)
plt.plot(T, y4)
plt.title('y4 = 0.5cos(2π6t + π)')
plt.grid(True)

plt.tight_layout()
plt.show() 
```
